```python
import pandas as pd
import os
```

```python
path = '5H1AI_logs/'
```

```python
columns = ['FILE','STATE','INDEX','ACTIONS','HANDS','FINAL','PLAYERS']
df = pd.DataFrame(columns=columns)
for f in os.listdir(path):
    current_file = path + str(f)
    cur_df = pd.read_csv(current_file,sep=':',skiprows=4,names=columns[1:])
    cur_df['FILE'] = f
    df = pd.concat([df,cur_df],axis=0,sort=True,ignore_index=True)
df = df.loc[:,columns]
df
```

```python
df.groupby('FILE').size()
```

```python
df['ACTIONS'] = df['ACTIONS'].str.split('/')
df['HANDS'] = df['HANDS'].str.split('|')
df['FINAL'] = df['FINAL'].str.split('|')
df['PLAYERS'] = df['PLAYERS'].str.split('|')
df
```

```python
df['PLURIBUS_SEAT'] = df['PLAYERS'].str.index('Pluribus')
df
```

```python
df['PLURIBUS_FINAL'] = pd.Series([float(x[0][x[1]]) for x in zip(df['FINAL'],df['PLURIBUS_SEAT'])])
df
```

```python
df['FLOP'] = [(x[5].split('/')[1:]+[0])[0] for x in df['HANDS']]
df['TURN'] = [(x[5].split('/')[2:]+[0])[0] for x in df['HANDS']]
df['RIVER'] = [(x[5].split('/')[3:]+[0])[0] for x in df['HANDS']]
df['HANDS'] = pd.Series([x[:5]+(x[5].split('/')[:1]) for x in df['HANDS']])
df
```

```python
interesting_hands = df.loc[(df['PLURIBUS_FINAL']>50)|(df['PLURIBUS_FINAL']<-100),:]
interesting_hands
```

```python
import re

def next_to_act(seat,remaining_players):
    for i in range(1,6):
        next_seat = (seat+i)%6
        if remaining_players[next_seat][1]==1:
            current_player = remaining_players[next_seat][0]
            cards = remaining_players[next_seat][2]
            return next_seat,current_player,cards

def to_poker_lingo(df):
    remaining_players = [[p[0],1,p[1]] for p in zip(df['PLAYERS'],df['HANDS'][:6])]
    for i,a in enumerate(df['ACTIONS']):
        seat=-1
        if i==0: seat=1 # preflop, start next to BB
        streets = ['PREFLOP:',f'FLOP: {df["FLOP"]}',f'TURN: {df["TURN"]}',f'RIVER:{df["RIVER"]}']
        print()
        print(streets[i])
        print()
        actions = list(filter(None,re.split('(f|c|r\d+)',a)))
        agg=False
        for j,b in enumerate(actions):
            seat,current_player,cards = next_to_act(seat,remaining_players)
            if b=="f":
                print(f'{current_player} ({cards}) FOLDS')
                remaining_players[seat][1]=0
            elif b=="c":
                if agg:
                   print(f'{current_player} ({cards}) CALLS')
                else:
                   print(f'{current_player} ({cards}) CHECKS')
            else:
                print(f'{current_player} ({cards}) RAISES {b[1:]}')
                agg=True
    print()
    print('RESULTS:')
    for i in zip(df['PLAYERS'],df['FINAL']):
        print(i[0],i[1])
    print('*'*20)
    print('*'*20)
```
